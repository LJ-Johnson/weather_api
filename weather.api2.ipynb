```python
LJ_key = #api key
baseurl = "http://api.weatherapi.com/v1/"
city = "Lexington"
```

```python
response = requests.get(f"http://api.weatherapi.com/v1/current.json?key={LJ_key}&q={city}")
response.status_code
```

```python
forecast_url = f"{baseurl}forecast.json?key={LJ_key}&q={city}&days=3"
forecast_response = requests.get(forecast_url)
forecast_response.status_code
```

```python
forecast_data = forecast_response.json()
forecast = forecast_data['forecast']['forecastday']
```

```python
forecast_3_days_summary = []
for day in forecast:
    forecast_3_days_summary.append(f"{day['date']}: High{day['day']['maxtemp_f']}f,Low {day['day']['mintemp_f']}f - {day['day']['condition']['text']}")
```

```python
forecast_3_days_summary = []
for day in forecast:
    date_obj = datetime.strptime(day['date'], '%Y-%m-%d')
    forematted_date = date_obj.strftime('%A, %B %#d')
    forecast_3_days_summary.append(f"{forematted_date}:High{day['day']['maxtemp_f']}f,Low {day['day']['mintemp_f']}f - {day['day']['condition']['text']}")

forecast_3_days_summary
```
